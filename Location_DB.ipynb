### Combining charts together and editing the combined table

```python
#combined charts together to be edited 
combined = [NJ_MID_boulders, NJ_SOUTH_boulders, PA_boulders, Delaware_Boulders]
combined_chart = pd.concat(combined)
combined_chart
```

```python
#checking the value in location
combined_chart.Location[1]
```

```python
#Spliting the values len to determine how many columns will need to be made
combined_chart.Location.str.split(' > ').apply(len).value_counts().sort_index()
```

```python
#Spliting the values at > to separate strings
location_df = combined_chart.Location.apply(lambda x: pd.Series(x.split(' > ')[::-1]))
location_df
```

```python
#Making columns to add in the separate values
combined_chart = combined_chart.drop(columns='Location')
combined_chart['State'] = location_df[0]
combined_chart['Gorge'] = location_df[1]
combined_chart['Area'] = location_df[2]
combined_chart['Crag'] = location_df[3]
combined_chart['Subcrag'] = location_df[4]
combined_chart.head()
```

### Convert table to SQL DB

```python
import sqlite3
```

```python
#Creating the database
conn = sqlite3.connect('boulder_DB.sqlite')
c = conn.cursor()
```

```python
#Creating a table and adding in the columns
c.execute('CREATE TABLE IF NOT EXISTS boulder_routes (Route text, Route_Type text, Grading text, Area_Latitude text, Area_Longitude text, State text, Gorge text, Area text, Crag text, Subcrag text, Description text, URL text)')
conn.commit()
```

```python
#Adding the chart to the database
combined_chart.to_sql('boulder_routes', conn, if_exists='replace', index = False)
```

```python
#Checking to see if the data was added in
c.execute('''  
SELECT * FROM boulder_routes
          ''')

for row in c.fetchall():
    print (row)
```

### Checking to see if the database was created and could be called back

```python
def qry(q, connection = sqlite3.connect("../PassionProject/boulder_DB.sqlite")):
    df = pd.read_sql_query(q, connection)
    connection.close
    return df
```

```python
tables = qry("""
    SELECT name FROM sqlite_master
""")

# Get Col + info
columns_in_table = pd.DataFrame()
for i in tables.name:
    df_i = qry("PRAGMA TABLE_INFO(" + i + ");")
    df_i['table_name'] = i
    columns_in_table = columns_in_table.append(df_i)
tables = tables.name
print(tables)
```
