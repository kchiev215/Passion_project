```python
import pandas as pd
import numpy as np
import csv
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
#specify the URL containing the dataset and pass it to urlopen() to get the html of the page.
url = "https://www.mountainproject.com/route-finder?diffMaxaid=75260&diffMaxboulder=21700&diffMaxice=38500&diffMaxmixed=60000&diffMaxrock=5500&diffMinaid=70000&diffMinboulder=20000&diffMinice=30000&diffMinmixed=50000&diffMinrock=1800&is_sport_climb=1&is_top_rope=1&is_trad_climb=1&pitches=0&selectedIds=106861605&sort1=area&sort2=rating&stars=0&type=boulder&viewAll=1"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
text = soup.get_text()
```

```python
soup.find_all('a')
```

```python
links = []
all_links = soup.find_all("a")
for link in all_links:
    links.append(link.get("href"))
    
links
```

```python
route_links=[]
for link in links:
    if 'https://www.mountainproject.com/route/' in link: #taking route only links to a new list
        route_links.append(link)
route_links
```

```python
len(route_links)
```

```python
route_links.index('https://www.mountainproject.com/route/120814193/good-dirty-fun')
```

```python
route_links = route_links[:55] #get the slice list from when the list repeats
```

```python
route_links
```

```python
len(route_links)
```

```python
end_of_url = [] #taking the ends of the link
for url in route_links:
    end_of_url.append(url.rsplit('/', 1)[-1])

print(end_of_url)
```

```python
end_of_url_df = pd.DataFrame(end_of_url, columns=['Route name'])
end_of_url_df['Route name'] = end_of_url_df['Route name'].str.replace("-", ' ').str.title()
end_of_url_df
```

```python
descriptions = [] 
for url in route_links:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    h1 = soup.find('div', class_= "fr-view") #descriptions are in this class

    descriptions.append(h1.get_text(strip=True)) #getting descriptions into text and stripping them
```

```python
descriptions
```

```python
end_of_url_df['Description'] = descriptions #adding description to dataframe
```

```python
end_of_url_df.head()
```

```python
end_of_url_df.to_csv('DE_description.csv') #transfer to csv
```
